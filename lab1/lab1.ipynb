### Функция потерь и градиентный спуск

```python
def error_fun(y, y_):
	diff = y - y_
	return np.sum(diff * diff) / len(y)

def grad_a(x, y, a, b):
    return 2 * np.sum(-x * (y - (a * x + b))) / len(x)

def grad_b(x, y, a, b):
    return  2 * np.sum(-1 * (y - (a * x + b))) / len(x)

def line_func(x, a, b):
	return a * x + b 
```

```python
a = np.random.randint(-10,10)
b = np.random.randint(-10,10)
step = 0.01
eps = 0.001

error = error_fun(y, line_func(x, a, b))
error_old = error + 1


history_gradient_err = [error]
history_grad_a = [a]
history_grad_b = [b]

while abs(error - error_old) > eps:
	a -= grad_a(x, y, history_grad_a[-1], history_grad_b[-1]) * step
	b -= grad_b(x, y, history_grad_a[-1], history_grad_b[-1]) * step
	error_old = error
	error = error_fun(y, line_func(x, a, b))
	history_gradient_err.append(error)
	history_grad_a.append(a)
	history_grad_b.append(b)

graf = make_subplots()
graf.add_trace(go.Scatter(x=history_gradient_err))
print('Оптимальные значения и значение ошибки')
print(a)
print(b)
print(error)
graf
```

```python
def normalize(x):
    x_min = np.min(x)
    x_max = np.max(x)

    return (x - x_min) / (x_max - x_min)
```

```python
with open("ex1data2.txt") as data:
    tmp_data = data.read().split('\n')[0:-1]
x1, x2, y = list(zip(*[sample.split(',') for sample in tmp_data]))
x1 = np.array([float(i) for i in x1])
x2 = np.array([float(i) for i in x2])
orig_y = np.array([float(i) for i in y])

x1 = np.concatenate([x1, [1650]])
x2 = np.concatenate([x2, [3]])

y = normalize(orig_y)
x = np.array([normalize(x1), normalize(x2)]).T
```

```python
x_test = x[len(x) - 1 : len(x)]
x = x[: len(x) - 1 ]
```

```python
def grad_theta(x, y, theta, index):
    return  2 * np.sum(-x[:, index] * (y - np.sum(theta * x))) / len(x)

def error_fun(y, y_):
    diff = y - y_
    return np.sum(diff * diff) / len(y)

def line_func(x, theta):
    return np.sum(x * theta) 
```

```python
theta = [np.random.randint(-2,2), np.random.randint(-2,2)]
step = 0.00001
eps = 0.0001
error = error_fun(y, line_func(x, theta))
error_old = error + 1

history_gradient_err = [error]
history_grad_theta = [theta]
count = 0

while abs(error - error_old) > eps:
    for i in range(len(theta)):
        theta[i] -= grad_theta(x, y, history_grad_theta[-1], i) * step
    error_old = error
    error = error_fun(y, line_func(x, theta))
    history_gradient_err.append(error)
    history_grad_theta.append(theta)
    count += 1

graf = make_subplots()
graf.add_trace(go.Scatter(y=history_gradient_err))
print(theta)
print(error)
graf
```
